```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
dataset= pd.read_csv("Train.csv")
dataset.head()
```

```python
X= pd.DataFrame(dataset.iloc[:,:-1])
X.head()
```

```python
y= pd.DataFrame(dataset.iloc[:,-1])
y.head()
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#building linear regression model
lr = LinearRegression().fit(X_train,y_train)
```

```python
#displaying slope and intercept values of regression line
print(lr.coef_)
print(lr.intercept_)
```

```python
test_dataframe= pd.read_csv("Test.csv")
test_dataframe.head()
```

```python
y_pred = lr.predict(test_dataframe)
y_pred = pd.DataFrame(y_pred, columns= ['Predicted'])
y_pred
```

```python
#finding error on train data
predictions_train = lr.predict(X_train)
error = mean_absolute_error(predictions_train,y_train)
print("MAE Train : ",error)
```

```python
#finding error on test data
predictions_test = lr.predict(X_test)
error = mean_absolute_error(predictions_test,y_test)
print("MAE Test : ",error)
```
